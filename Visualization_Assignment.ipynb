4.  Using the given x and y data, generate a 3D surface plot to visualize the function sin(sqrt(x**2 + y**2))

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create the 3D surface plot using Plotly
fig = go.Figure()
fig.add_trace(go.Surface(x = x, y = y, z = z))

# Customize the plot
fig.update_layout(title='3D Surface Plot of sin(sqrt(x^2 + y^2))',
                  autosize=False,
                  width=800,
                  height=600,
                  scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'))

fig.show()
```

5. Create a Bokeh heatmap using the provided dataset. 

```python
import numpy as Numpy
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.palettes import Viridis256

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

x_flat = xx.flatten()
y_flat = yy.flatten()
values = data_heatmap.flatten()

mapper = linear_cmap(field_name = 'value',palette = Viridis256, low = min(values),high = max(values))

p = figure(title = "Heatmap", x_axis_label = 'X', y_axis_label = 'Y', width = 600, height = 600)
p.rect(x = 'x', y = 'y', width = 0.1, height = 0.1, source = dict(x = x_flat, y = y_flat, value = values),fill_color=mapper,line_color=None)

color_bar = ColorBar(color_mapper = mapper['transform'], width = 8, location = (0,0))
p.add_layout(color_bar,'right')

output_notebook()
show(p)
```
